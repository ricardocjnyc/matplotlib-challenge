```python
Capomulin_volume_initial=all_data_df_by_drug_timepoint_volume_group["Capomulin"][0]
Capomulin_volume_last=all_data_df_by_drug_timepoint_volume_group["Capomulin"][45]
change_Capomulin_volume=-(1-Capomulin_volume_last/Capomulin_volume_initial)
Infubinol_volume_initial=all_data_df_by_drug_timepoint_volume_group["Infubinol"][0]
Infubinol_volume_last=all_data_df_by_drug_timepoint_volume_group["Infubinol"][45]
change_Infubinol_volume=-(1-Infubinol_volume_last/Infubinol_volume_initial)
Capomulin_volume_initial=all_data_df_by_drug_timepoint_volume_group["Capomulin"][0]
Capomulin_volume_last=all_data_df_by_drug_timepoint_volume_group["Capomulin"][45]
change_Capomulin_volume=-(1-Capomulin_volume_last/Capomulin_volume_initial)
Ketapril_volume_initial=all_data_df_by_drug_timepoint_volume_group["Ketapril"][0]
Ketapril_volume_last=all_data_df_by_drug_timepoint_volume_group["Ketapril"][45]
change_Ketapril_volume=-(1-Ketapril_volume_last/Infubinol_volume_initial)
Placebo_volume_initial=all_data_df_by_drug_timepoint_volume_group["Placebo"][0]
Placebo_volume_last=all_data_df_by_drug_timepoint_volume_group["Placebo"][45]
change_Placebo_volume=-(1-Placebo_volume_last/Placebo_volume_initial)
```

```python
volume_change_list=(change_Capomulin_volume,change_Infubinol_volume,change_Ketapril_volume,change_Placebo_volume)
drug_list=("Capomulin","Infubinol","Ketapril","Placebo")
drugs_grade=["True","False"]
```

```python
volumes_change={"Volume change":[change_Capomulin_volume,change_Infubinol_volume,
                                change_Ketapril_volume,change_Placebo_volume],
                "Drug":["Capomulin","Infubinol","Ketapril","Placebo"]
               }      
volumes_change_pf=pd.DataFrame(volumes_change, columns=["Drug","Volume change"])
```

```python
bins=[-1,0,1]
```

```python
volumes_change_pf["Result_of_test"]=pd.cut(volumes_change_pf["Volume change"], bins, labels=drugs_grade).head()
volumes_change_pf
```

```python
drug_groups=volumes_change_pf.groupby(["Result_of_test"])
```

```python
Capomulin_result=volumes_change_pf["Result_of_test"][0]
Infubinol_result=volumes_change_pf["Result_of_test"][1]
Ketapril_result=volumes_change_pf["Result_of_test"][2]
Placebo_result=volumes_change_pf["Result_of_test"][3]
```

```python
if Capomulin_result=="False":
    color_Copomulin='r'
else:
    color_Copomulin='g'
```

```python
if Infubinol_result=="False":
    color_Infubinol='r'
else:
    color_Infubinol='g'
```

```python
if Ketapril_result=="False":
    color_Ketapril='r'
else:
    color_Ketapril='g'
```

```python
if Placebo_result=="False":
    color_Placebo='r'
else:
    color_Placebo='g'
```

```python
plt.title("Tumor change over 45 day treatment")
plt.ylabel("% Tumor volume change")
bar1=plt.bar("Capomulim",change_Capomulin_volume*100, color=color_Copomulin)
bar2=plt.bar("Infubinol",change_Infubinol_volume*100, color=color_Infubinol)
bar3=plt.bar("Ketapril",change_Ketapril_volume*100, color=color_Ketapril)
bar4=plt.bar("Placebo",change_Placebo_volume*100, color=color_Placebo)


def autolabel(rects):
    width = 0.35  
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2.0, height, '%d ' % int(height), ha='center', va='bottom')    

autolabel(bar1)
autolabel(bar2)
autolabel(bar3)
autolabel(bar4)
        
plt.tight_layout()

plt.grid()
plt.show()
```
